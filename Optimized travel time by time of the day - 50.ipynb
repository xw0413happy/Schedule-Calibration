```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime
from datetime import timedelta
```

```python
# define a function that returns optimized segment time by hours

def seg_time_hourly(opt_seg_time_file):
    
    df_times = pd.DataFrame()
    df = pd.read_csv(opt_seg_time_file, parse_dates =["TripStartTime"], index_col ="TripStartTime") # Direction 1

    seg = df['Segment'].tolist()
    seg_unic = list(set(seg))

    print(seg_unic)

    for seg in seg_unic:
        seg_df = df.loc[df['Segment'] == seg]
        resampled_seg_df = seg_df.resample('60min').mean()
        resampled_seg_df['segment'] = seg
        del resampled_seg_df["Unnamed: 0"]
        # # remove the dates from the field
        resampled_seg_df.reset_index(inplace=True)
        resampled_seg_df['TIME_INTERVALS'] = resampled_seg_df['TripStartTime'].dt.time
        del resampled_seg_df["TripStartTime"]
        resampled_seg_df = resampled_seg_df.interpolate(method='linear', axis=0).ffill().bfill()
        resampled_seg_df[['seg_start','seg_end']] = resampled_seg_df.segment.str.split(" - ", expand=True)
        resampled_seg_df = resampled_seg_df[['TIME_INTERVALS', 'seg_start', 'seg_end', 'segment', 'CURRENT_SCHED_TIME', 'OPTM_SCHED_TIME']]
        df_times = df_times.append(resampled_seg_df, ignore_index = True)
    df_times.to_csv(opt_seg_time_file[:-4] + '_travel_times.csv')
    return df_times   
```

# Generate the travel time reference table for scheduler

```python
# Weekdays
direct_1_opt_seg_time = 'Calibration_50 EAST Weekdays - Dec - Mar_Brief.csv'
direct_2_opt_seg_time = 'Calibration_50 WEST Weekdays - Dec - Mar_Brief.csv'

df1_times = seg_time_hourly(direct_1_opt_seg_time)
df2_times = seg_time_hourly(direct_2_opt_seg_time)

df1_times.to_csv('Calibrated_50 EAST Weekdays_travel times.csv')
df2_times.to_csv('Calibrated_50 WEST Weekdays_travel times.csv')

# Saturdays
direct_1_opt_seg_time = 'Calibration_50 EAST Saturdays - Dec - Mar_Brief.csv'
direct_2_opt_seg_time = 'Calibration_50 WEST Saturdays - Dec - Mar_Brief.csv'

df1_times = seg_time_hourly(direct_1_opt_seg_time)
df2_times = seg_time_hourly(direct_2_opt_seg_time)

df1_times.to_csv('Calibrated_50 EAST Saturdays_travel times.csv')
df2_times.to_csv('Calibrated_50 WEST Saturdays_travel times.csv')

# Sundays
direct_1_opt_seg_time = 'Calibration_50 EAST Sundays - Dec - Mar_Brief.csv'
direct_2_opt_seg_time = 'Calibration_50 WEST Sundays - Dec - Mar_Brief.csv'

df1_times = seg_time_hourly(direct_1_opt_seg_time)
df2_times = seg_time_hourly(direct_2_opt_seg_time)

df1_times.to_csv('Calibrated_50 EAST Sundays_travel times.csv')
df2_times.to_csv('Calibrated_50 WEST Sundays_travel times.csv')

print('Done')
```

```python
direct_1_opt_seg_time = 'Calibration_50 EAST Sundays - Dec - Mar_Brief.csv'
direct_2_opt_seg_time = 'Calibration_50 WEST Sundays - Dec - Mar_Brief.csv'

df1_times = seg_time_hourly(direct_1_opt_seg_time)
df2_times = seg_time_hourly(direct_2_opt_seg_time)

df1_times.to_csv('Route 50 EAST Sundays_time.csv')
df2_times.to_csv('Route 50 WEST Sundays_time.csv')

drect_1_start_time = '07:20:00'

layover1 = '00:00:00'
layover2 = '00:10:00'
input_stop_order1 = ['bearid', 'sanout', 'heapar', 'winsum', 'btwrsh', 'dantre', 'rsw']
input_stop_order2 = ['rsw', 'dantre', 'btwrsh', 'winsum', 'heapar', 'bearid']

number_of_trips_one_block = 9

#############################################################

start_time = drect_1_start_time
input_stop_order = input_stop_order1

passing_times = [start_time]
passing_times_h = start_time[:2] + ':00:00' # calculate the hour of the start hour


trip_id = 1
while trip_id <= number_of_trips_one_block:
    
    print('Trip:', trip_id)
    print('--Direction 1')
    print(input_stop_order)
    print (*input_stop_order, sep =', ')
    for i in range(len(input_stop_order)-1):

        passing_times_h = datetime.strptime(passing_times_h, '%H:%M:%S') # convert string to datetime (date + time)
        passing_times_h = passing_times_h.time() # keep the time and remove dates

        opt_time = df1_times.loc[(df1_times['TIME_INTERVALS'] == passing_times_h) 
                                 & (df1_times['seg_start']== input_stop_order[i])]

        opt_time = opt_time['OPTM_SCHED_TIME'].values[0] # find the segment time to use for the trip

        start_time_calcu = datetime.strptime(passing_times[i], '%H:%M:%S') # convert the start time into datetime
        b = start_time_calcu + timedelta(seconds = int(opt_time)) # add the optimized seg travel time to the time
        c = str(b)[11:]
        passing_times.insert(i+1, c)
        passing_times_h = start_time[:2] + ':00:00' # calculate the hour of the start hour
        
    print(passing_times)
    print (*passing_times, sep =', ')
    start_time_calcu = datetime.strptime(c, '%H:%M:%S') + timedelta(hours = int(layover1[:2]), 
                                                                    minutes = int(layover1[3:5]),
                                                                    seconds = int(layover1[6:8]))
    print()
    
    print('--direction 2')
    start_time = start_time_calcu.time()
    start_time = str(start_time)
    input_stop_order = input_stop_order2
    print(input_stop_order)
    print (*input_stop_order, sep =', ')
    passing_times = [start_time]
    passing_times_h = start_time[:2] + ':00:00' # calculate the hour of the start hour
    
    for i in range(len(input_stop_order)-1):

        passing_times_h = datetime.strptime(passing_times_h, '%H:%M:%S') # convert string to datetime (date + time)
        passing_times_h = passing_times_h.time() # keep the time and remove dates

        opt_time = df2_times.loc[(df2_times['TIME_INTERVALS'] == passing_times_h) 
                                   & (df2_times['seg_start']== input_stop_order[i])]
        opt_time = opt_time['OPTM_SCHED_TIME'].values[0] # find the segment time to use for the trip

        start_time_calcu = datetime.strptime(passing_times[i], '%H:%M:%S') # convert the start time into datetime
        b = start_time_calcu + timedelta(seconds = int(opt_time)) # add the optimized seg travel time to the time
        c = str(b)[11:]
        passing_times.insert(i+1, c)
        passing_times_h = start_time[:2] + ':00:00' # calculate the hour of the start hour

    print(passing_times)
    print (*passing_times, sep =', ')
    start_time_calcu = datetime.strptime(c, '%H:%M:%S') + timedelta(hours = int(layover2[:2]), 
                                                                minutes = int(layover2[3:5]),
                                                                seconds = int(layover2[6:8]))
    
    start_time = start_time_calcu.time()
    start_time = str(start_time)
    input_stop_order = input_stop_order1
    passing_times = [start_time]
    passing_times_h = start_time[:2] + ':00:00' # calculate the hour of the start hour
    
    print('')
    print('')
    trip_id = trip_id + 1
```
